```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
dataset = pd.read_csv('C:/Users/Maalaoui/Desktop/h_p/all_perth_310121.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
#GARAGE
#BUILD_YEAR
#NEAREST_SCH_RANK
#3 col with null values
#to replace with mean value
```

```python
dataset['GARAGE']=dataset['GARAGE'].fillna(dataset['GARAGE'].mean())
dataset['BUILD_YEAR']=dataset['BUILD_YEAR'].fillna(dataset['BUILD_YEAR'].mean())
dataset['NEAREST_SCH_RANK']=dataset['NEAREST_SCH_RANK'].fillna(dataset['NEAREST_SCH_RANK'].mean())
```

```python
dataset.info()
```

```python
le = LabelEncoder()

x1=dataset['ADDRESS']
x1new = le.fit_transform(x1)
dataset['ADDRESS'] =x1new.reshape(-1,1)

x2=dataset['SUBURB']
x2new = le.fit_transform(x2)
dataset['SUBURB'] =x2new.reshape(-1,1)

x3=dataset['NEAREST_STN']
x3new = le.fit_transform(x3)
dataset['NEAREST_STN'] =x3new.reshape(-1,1)

x4=dataset['NEAREST_SCH']
x4new = le.fit_transform(x4)
dataset['NEAREST_SCH'] =x4new.reshape(-1,1)
```

```python
dataset["DATE_SOLD"] = pd.to_datetime(dataset["DATE_SOLD"])
dataset["SOLD_YEAR"] = dataset["DATE_SOLD"].dt.year
dataset["SOLD_MONTH"] = dataset["DATE_SOLD"].dt.month
```

```python
label= np.array(dataset['PRICE'])
dataset = dataset.drop(['DATE_SOLD','PRICE'],axis=1)
```

```python
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, train_label, test_label = train_test_split(dataset, label, test_size =0.33,random_state=0) 
```

```python
###############RandomForest###############
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=0,n_estimators=50).fit(x_train,train_label)
pred=model.predict(x_test)
```

```python
from sklearn.metrics import explained_variance_score
ev=explained_variance_score(test_label, pred)
print('explained_variance:\n')
print(ev)
```

```python
###############xgboost###############
```

```python
import xgboost as xgb
model = xgb.XGBRegressor(random_state=0,n_estimators=80,max_depth=5).fit(x_train,train_label)
predxgb=model.predict(x_test)
```

```python
from sklearn.metrics import explained_variance_score
ev=explained_variance_score(test_label, predxgb)
print('explained_variance:\n')
print(ev)
```
